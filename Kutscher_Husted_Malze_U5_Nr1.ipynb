## Aufgabe 1 (a)

Zum Vergleich der Performance des UltraSort Algorithmus wurde auch ein Durchlauf nur mit InsertionSort gemacht, und den erreichten Werten der 1/3 Fraktionierung entgegen gestellt.

```python
from matplotlib import pyplot as plt

data_size = [10000, 100000, 1000000]

random_1to3 = [0.070513, 5.89945, 564.9]
rise_1to3 =   [0.001481, 0.012577, 0.14155]
fall_1to3 =   [0.130946, 11.2934, 1132.9]

rand_Ins =    [0.329991, 28.853, 2830.45]
rise_Ins =    [0.000151, 0.001312, 0.013063]
fall_Ins =    [0.650946, 57.2774, 5600]
```

```python

plt.plot(data_size, random_1to3, color='g', label='Zufällig 1/3')
plt.plot(data_size, rand_Ins,color='g', label='Zufällig insert', linestyle='-.')
plt.xlabel('Datengröße in Elemente')
plt.ylabel('Laufzeit in Sekunden')
plt.title('Laufzeit InsertionSort und Ultrasort 1/3 vs Datengröße')
plt.legend()
```

```python
plt.plot(data_size, rise_1to3, color='r', label='Aufsteigend 1/3')
plt.plot(data_size, rise_Ins,color='r', label='Aufsteigend insert', linestyle='-.')
plt.xlabel('Datengröße in Elemente')
plt.ylabel('Laufzeit in Sekunden')
plt.title('Laufzeit InsertionSort und Ultrasort 1/3 vs Datengröße')
plt.legend()
```

```python
plt.plot(data_size, fall_1to3, color='b', label='Absteigend 1/3')
plt.plot(data_size, fall_Ins,color='b', label='Absteigend insert', linestyle='-.')
plt.ylabel('Laufzeit in Sekunden')
plt.title('Laufzeit InsertionSort und Ultrasort 1/3 vs Datengröße')
plt.legend()
```

Die Laufzeiten steigen mit ansteigendet Datengröße linear an. Im Vergleich zu Insertion Sort ist der Ultra Sort Algorithmus für zufällige und absteignde Listen
schneller. Im Falle der aufsteigenden Listen ist die Laufzeit für Ultra sort deutlich schlechter.
